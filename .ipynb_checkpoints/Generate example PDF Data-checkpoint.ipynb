```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.lib.colors import black, red, blue, green, orange, purple
from reportlab.lib.utils import ImageReader
from io import BytesIO
from PIL import Image, ImageDraw, ImageFont
import random

def create_image_with_text(text: str, width: int, height: int) -> Image.Image:
    """Create an image with the given text."""
    image = Image.new('RGB', (width, height), color='white')
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()
    draw.text((10, 10), text, font=font, fill='black')
    return image

def add_colored_border(pdf, color):
    """Add a colored border to the page."""
    pdf.setStrokeColor(color)
    pdf.setLineWidth(10)
    pdf.rect(5, 5, letter[0]-10, letter[1]-10, fill=0, stroke=1)

def add_watermark(pdf, text):
    """Add a watermark to the page."""
    pdf.saveState()
    pdf.setFont("Helvetica", 60)
    pdf.setFillColor(blue, alpha=0.3)
    pdf.translate(letter[0]/2, letter[1]/2)
    pdf.rotate(45)
    pdf.drawCentredString(0, 0, text)
    pdf.restoreState()

def add_header(pdf, text, color):
    """Add a colored header to the page."""
    pdf.setFillColor(color)
    pdf.rect(0, letter[1] - 0.5*inch, letter[0], 0.5*inch, fill=1, stroke=0)
    pdf.setFillColor(black)
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(0.5*inch, letter[1] - 0.3*inch, text)

def add_footer(pdf, text, color):
    """Add a colored footer to the page."""
    pdf.setFillColor(color)
    pdf.rect(0, 0, letter[0], 0.5*inch, fill=1, stroke=0)
    pdf.setFillColor(black)
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(0.5*inch, 0.2*inch, text)

def create_rotated_text_pdf(filename: str) -> None:
    """
    Create a PDF with 15 pages, visually grouped into five documents.
    :param filename: str, the name of the output PDF file
    """
    pdf = canvas.Canvas(filename, pagesize=letter)
    width, height = letter

    # Document 1: Pages 1-3 (Colored border)
    for page in range(3):
        add_colored_border(pdf, red)
        
        if page == 0:
            # First page with logo
            logo_path = "./logo.png"
            pdf.drawImage(logo_path, 0.5*inch, 0.5*inch, width=width-inch, height=height-inch, preserveAspectRatio=True, anchor='c')
        else:
            text = f"This is page {page + 1} of Document 1"
            rotation = random.randint(0, 359)
            print(rotation)
            pdf.setFont("Helvetica", 14)
            pdf.setFillColor(black)
            x, y = width / 2, height / 2
            pdf.saveState()
            pdf.translate(x, y)
            pdf.rotate(rotation)
            pdf.drawString(-100, 0, text)
            pdf.restoreState()
        
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 1 - Page {page + 1}")
        pdf.showPage()

    # Document 2: Pages 4-6 (Watermark)
    for page in range(3):
        add_watermark(pdf, "Document 2")
        
        text = f"This is page {page + 1} of Document 2"
        rotation = random.randint(0, 359)
        print(rotation)
        img = create_image_with_text(text, int(width), int(height))
        img = img.rotate(rotation, expand=1)
        img_reader = ImageReader(img)
        pdf.drawImage(img_reader, 0, 0, width=width, height=height)
        
        pdf.setFillColor(black)
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 2 - Page {page + 1}")
        pdf.showPage()

    # Document 3: Pages 7-9 (Colored background)
    for page in range(3):
        pdf.setFillColor(green, alpha=0.1)
        pdf.rect(0, 0, width, height, fill=1, stroke=0)
        
        text = f"This is page {page + 1} of Document 3"
        rotation = random.randint(0, 359)
        print(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 3 - Page {page + 1}")
        pdf.showPage()

    # Document 4: Pages 10-12 (Colored header)
    for page in range(3):
        add_header(pdf, f"Document 4 - Page {page + 1}", orange)
        
        text = f"This is page {page + 1} of Document 4"
        rotation = random.randint(0, 359)
        print(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.showPage()

    # Document 5: Pages 13-15 (Colored footer)
    for page in range(3):
        add_footer(pdf, f"Document 5 - Page {page + 1}", purple)
        
        text = f"This is page {page + 1} of Document 5"
        rotation = random.randint(0, 359)
        print(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.showPage()
        
    # Document 6: Pages 16-18 (Watermark)
    for page in range(3):
        add_watermark(pdf, "Document 2")
        
        text = f"This is page {page + 1} of Document 2"
        rotation = random.randint(0, 359)
        print(rotation)
        img = create_image_with_text(text, int(width), int(height))
        img = img.rotate(rotation, expand=1)
        img_reader = ImageReader(img)
        pdf.drawImage(img_reader, 0, 0, width=width, height=height)
        
        pdf.setFillColor(black)
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 2 - Page {page + 1}")
        pdf.showPage()

    pdf.save()

output_pdf = "grouped_documents.pdf"
create_rotated_text_pdf(output_pdf)
print(f"PDF '{output_pdf}' has been created with 18 pages grouped into five visually distinct documents.")
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.lib.colors import black, red, blue, green, orange, purple
from reportlab.lib.utils import ImageReader
from PyPDF2 import PdfReader, PdfWriter
from pdf2image import convert_from_path
import tempfile
from io import BytesIO
from PIL import Image, ImageDraw, ImageFont
import random
import os

def create_image_with_text(text: str, width: int, height: int) -> Image.Image:
    """Create an image with the given text."""
    image = Image.new('RGB', (width, height), color='white')
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()
    draw.text((10, 10), text, font=font, fill='black')
    return image

def add_colored_border(pdf, color):
    """Add a colored border to the page."""
    pdf.setStrokeColor(color)
    pdf.setLineWidth(10)
    pdf.rect(5, 5, letter[0]-10, letter[1]-10, fill=0, stroke=1)

def add_watermark(pdf, text):
    """Add a watermark to the page."""
    pdf.saveState()
    pdf.setFont("Helvetica", 60)
    pdf.setFillColor(blue, alpha=0.3)
    pdf.translate(letter[0]/2, letter[1]/2)
    pdf.rotate(45)
    pdf.drawCentredString(0, 0, text)
    pdf.restoreState()

def add_header(pdf, text, color):
    """Add a colored header to the page."""
    pdf.setFillColor(color)
    pdf.rect(0, letter[1] - 0.5*inch, letter[0], 0.5*inch, fill=1, stroke=0)
    pdf.setFillColor(black)
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(0.5*inch, letter[1] - 0.3*inch, text)

def add_footer(pdf, text, color):
    """Add a colored footer to the page."""
    pdf.setFillColor(color)
    pdf.rect(0, 0, letter[0], 0.5*inch, fill=1, stroke=0)
    pdf.setFillColor(black)
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(0.5*inch, 0.2*inch, text)

def create_rotated_text_pdf(filename: str):
    """
    Create a PDF with 15 pages, visually grouped into five documents.
    :param filename: str, the name of the output PDF file
    """

    rot_list = []
    
    pdf = canvas.Canvas(filename, pagesize=letter)
    width, height = letter

    # Document 1: Pages 1-3 (Colored border)
    for page in range(3):
        add_colored_border(pdf, red)
        
        if page == 0:
            # First page with logo
            logo_path = "./logo.png"
            pdf.drawImage(logo_path, 0.5*inch, 0.5*inch, width=width-inch, height=height-inch, preserveAspectRatio=True, anchor='c')
        else:
            text = f"This is page {page + 1} of Document 1"
            rotation = random.randint(0, 359)
            rot_list.append(rotation)
            pdf.setFont("Helvetica", 14)
            pdf.setFillColor(black)
            x, y = width / 2, height / 2
            pdf.saveState()
            pdf.translate(x, y)
            pdf.rotate(rotation)
            pdf.drawString(-100, 0, text)
            pdf.restoreState()
        
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 1 - Page {page + 1}")
        pdf.showPage()

    # Document 2: Pages 4-6 (Watermark)
    for page in range(3):
        add_watermark(pdf, "Document 2")
        
        text = f"This is page {page + 1} of Document 2"
        rotation = random.randint(0, 359)
        rot_list.append(rotation)
        img = create_image_with_text(text, int(width), int(height))
        img = img.rotate(rotation, expand=1)
        img_reader = ImageReader(img)
        pdf.drawImage(img_reader, 0, 0, width=width, height=height)
        
        pdf.setFillColor(black)
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 2 - Page {page + 1}")
        pdf.showPage()

    # Document 3: Pages 7-9 (Colored background)
    for page in range(3):
        pdf.setFillColor(green, alpha=0.1)
        pdf.rect(0, 0, width, height, fill=1, stroke=0)
        
        text = f"This is page {page + 1} of Document 3"
        rotation = random.randint(0, 359)
        rot_list.append(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 3 - Page {page + 1}")
        pdf.showPage()

    # Document 4: Pages 10-12 (Colored header)
    for page in range(3):
        add_header(pdf, f"Document 4 - Page {page + 1}", orange)
        
        text = f"This is page {page + 1} of Document 4"
        rotation = random.randint(0, 359)
        rot_list.append(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.showPage()

    # Document 5: Pages 13-15 (Colored footer)
    for page in range(3):
        add_footer(pdf, f"Document 5 - Page {page + 1}", purple)
        
        text = f"This is page {page + 1} of Document 5"
        rotation = random.randint(0, 359)
        rot_list.append(rotation)
        pdf.setFont("Helvetica", 14)
        pdf.setFillColor(black)
        x, y = width / 2, height / 2
        pdf.saveState()
        pdf.translate(x, y)
        pdf.rotate(rotation)
        pdf.drawString(-100, 0, text)
        pdf.restoreState()
        
        pdf.showPage()
        
    # Document 6: Pages 16-18 (Watermark)
    for page in range(3):
        add_watermark(pdf, "Document 2")
        
        text = f"This is page {page + 1} of Document 2"
        rotation = random.randint(0, 359)
        rot_list.append(rotation)
        img = create_image_with_text(text, int(width), int(height))
        img = img.rotate(rotation, expand=1)
        img_reader = ImageReader(img)
        pdf.drawImage(img_reader, 0, 0, width=width, height=height)
        
        pdf.setFillColor(black)
        pdf.drawString(0.5*inch, 0.25*inch, f"Document 2 - Page {page + 1}")
        pdf.showPage()

    pdf.save()
    return rot_list

def convert_pdf_page_to_image(page: 'PdfReader.PageObject') -> 'Image':
    """
    Convert a PDF page to an image.

    Args:
    page (PdfReader.PageObject): A single page from a PDF.

    Returns:
    Image: The converted image.
    """
    pdf_writer = PdfWriter()
    pdf_writer.add_page(page)
    
    # Save the single-page PDF to a temporary file
    with tempfile.NamedTemporaryFile(suffix=".pdf", delete=False) as temp_pdf:
        pdf_writer.write(temp_pdf)
        temp_pdf_path = temp_pdf.name
    
    # Convert the single-page PDF to an image using pdf2image with higher resolution
    images = convert_from_path(temp_pdf_path, dpi=300)
    
    # Clean up the temporary file
    os.remove(temp_pdf_path)
    
    return images[0]  # There should be only one page, hence one image

def convert_and_save(input_page, rotation) -> str:
    
    image = convert_pdf_page_to_image(input_page)

    path = 'C:/Users/arues/Documents/GitHub/hiringstudy01/training_rotation/val/' + str(rotation)

    #print(path)

    # Ensure the directory exists
    os.makedirs(path, exist_ok=True)
    
    with tempfile.NamedTemporaryFile(suffix=".png", dir=path, delete=False) as training_img:
        image.save(training_img)
        img_name = training_img.name
        pass
    return img_name

def save_pdf_to_img(input_pdf: str, rot_list: [int]) -> None:
    """
    Analyze all pages in the input PDF and determine the rotation angle needed for each page.

    Args:
    input_pdf (str): The file path of the input PDF.

    Returns:
    List[int]: A list of rotation angles (in degrees) for each page. 
               The angles are normalized to be in the range [0, 359].
               0 means no rotation needed, 90 means 90 degrees clockwise, etc.
    """
    reader = PdfReader(input_pdf)
    for page_number in range(len(reader.pages)):
        if page_number == 0:
            print(page_number)
        else:
            current_page = reader.pages[page_number]
            convert_and_save(current_page, rot_list[page_number-1])


def list_folders_in_path(path):
    try:
        # List all entries in the specified path
        entries = os.listdir(path)
        
        # Filter out entries that are directories
        folders = [entry for entry in entries if os.path.isdir(os.path.join(path, entry))]
        
        # Get the count of directories
        folder_count = len(folders)
        
        return folder_count
    except FileNotFoundError:
        print(f"The path {path} does not exist.")
        return 0
    except Exception as e:
        print(f"An error occurred: {e}")
        return 0


folder_count=list_folders_in_path('C:/Users/arues/Documents/GitHub/hiringstudy01/training_rotation/val/')
output_pdf = "grouped_documents.pdf"

#while folder_count < 360:
rot_list = create_rotated_text_pdf(output_pdf)
print(rot_list)
    #save_pdf_to_img(output_pdf, rot_list)
    #folder_count=list_folders_in_path('C:/Users/arues/Documents/GitHub/hiringstudy01/training_rotation/val/')
    #print(str(folder_count) + ' of 360 folders created. That is: ' + str(folder_count/360*100) + '%')

print(f"PDF '{output_pdf}' has been created with 18 pages grouped into five visually distinct documents.")
```
